# **Detecção de Faces**

**Classificador Cascade**<br>
Para esse classificador, no caso da detecção de faces, precisamos de um conjunto de imagens de faces e outro de não faces. O treinamento é feito com o AdaBoost, cuja seleção das características das imagens é feita a partir da aplicação de moldes em cada janela da imagem.
$$\sum Pixel_{Branco} - \sum Pixel_{Preto}$$
Existem mais de 160.000 combinações de pixels brancos e pretos em uma imagem 24x24.

## Faces

```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread('/Users/gustavomeira/Documents/Python/Estudo_ML/projects/Visão Computacional/workplace-1245776_1920.jpg')
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
plt.axis('off')
plt.show()
```

```python
detector_face = cv2.CascadeClassifier('/Users/gustavomeira/Documents/Python/Estudo_ML/projects/Visão Computacional/haarcascade_frontalface_default.xml')
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
plt.imshow(imagem_cinza)
plt.axis('off')
plt.show()
```

```python
deteccoes = detector_face.detectMultiScale(imagem_cinza, scaleFactor=1.3, minSize=(30,30))
```

```python
display(deteccoes)
```

```python
for (x,y,l,a) in deteccoes:
    cv2.rectangle(imagem, (x,y), (x+l,y+a), (0,255,0), 3)
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
plt.axis('off')
plt.show()
```

## Full Body

```python
imagem = cv2.imread('/Users/gustavomeira/Documents/Python/Estudo_ML/projects/Visão Computacional/pessoas.jpg')
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
plt.axis('off')
plt.show()
```

```python
detector_face = cv2.CascadeClassifier('/Users/gustavomeira/Documents/Python/Estudo_ML/projects/Visão Computacional/fullbody.xml')
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
plt.imshow(imagem_cinza)
plt.axis('off')
plt.show()
```

```python
deteccoes = detector_face.detectMultiScale(imagem_cinza, scaleFactor=1.1, minSize=(50,50))

for (x,y,l,a) in deteccoes:
    cv2.rectangle(imagem, (x,y), (x+l,y+a), (255,255,80), 2)
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
plt.axis('off')
plt.show()
```
